<h2 style="background-color: rgb(155,196,226); color: rgb(8,69,126); padding: 10px;">Part 5. Standardize binary / yes-no style columns</h2>
<p>This part is incredibly important.  Our data for stats analysis are going to be &ldquo;NOIR&rdquo; - nominal, ordinal, interval, ratio.  Some stats tests require nominal data, others binary, some columns may need to be cast to a different data type ... "  Computers like consistency: as you know, people who enter the data may be 'y', 'Y', 'Yes', "Ye", or 1 for "yes" - an int of 1 or a string of "Y" won't work together.  
We are converting all of those to one consistent system.</p>

```python
binary_cols = [
    "teacher_resources",
    "k_12",
    "university",
    "social_groups",
    "official_support",
    "media",
    "digital_resources",
    "volunteer_programs"
]

def clean_binary(value):
    if pd.isna(value):
        return np.nan
    
    value = str(value).strip().lower()
    
    yes_values = {"yes", "y", "1", "true", "t"}
    no_values = {"no", "n", "0", "false", "f"}
    
    if value in yes_values:
        return 1
    elif value in no_values:
        return 0
    else:
        return np.nan

for col in binary_cols:
    if col in df.columns:
        print(f"\nBefore cleaning: {col}")
        print(df[col].value_counts(dropna=False).head(10))
        df[col] = df[col].apply(clean_binary)
        print(f"After cleaning: {col}")
        print(df[col].value_counts(dropna=False))
```

<p>On a tangent, when you have a bunch of people entering data you can measure 
the degree of consistency among the inputters, called <i>inter-indexer consistency</i>.  There's Cohen's &kappa; or Krippendorfer's &alpha; as stat measures to get an idea of amount of consistency.</p>

<hr />
<h2 style="background-color: rgb(155,196,226); color: rgb(8,69,126); padding: 10px;">Part 6. Clean numeric columns stored as text</h2>
<p>And speaking of text ... are there fields that should be converted to text?   Or not?  How might we handle errors?  Why?  Recall data sent over the net or misencoded can look numeric but are actually strings.</p>
<p>Let's explore the <code>cost_per_annum</code></p>


```python
df["cost_per_annum"] = (
    df["cost_per_annum"]
    .astype(str)
    .str.replace(r"[\$,]", "", regex=True)
    .str.replace("USD", "", regex=False)
    .str.strip()
)

df["cost_per_annum"] = pd.to_numeric(df["cost_per_annum"], errors="coerce")
display(df["cost_per_annum"])
```

```python
# and always keep up to date with your data!  Inspect ... 
df["cost_per_annum"].describe()
```

<hr />
<h2  style="background-color: rgb(155,196,226); color: rgb(8,69,126); padding: 10px;">Part 7. Handle missing values</h2>
<p>If you've never done any interpolation you might want to avoid it.
But it is common enough when exploring teaching time-series logic.  
    Median/mode filling is easier to justify here. ('cause we're not a stats
    class and not stressing over error (see the Planning Weeks 9-14 doc).</p>
<p>For numeric columns, median is often a safe fill value.
For text columns, "unknown" can be better than leaving blanks.   [But note, even if you choose "unknown" make sure <i>all</i> the unknowns are the same (even if you choose to put in "NA" (not available") make sure it's a consistent "NA" and not "na", "nope", "nada", and definitely not "?").  
There is no single correct method; what matters is choosing a method and explaining it.   [You'll see this again in stats as part of "explaining the error."</p>

```python
numeric_cols = df.select_dtypes(include=["number"]).columns
categorical_cols = df.select_dtypes(exclude=["number"]).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

for col in categorical_cols:
    df[col] = df[col].fillna("unknown")


print(f"Numeric Columns: {numeric_cols}")
print(f"Categorical Columns: {categorical_cols}")
```

```python
# yup, verify ... 
print(df.isna().sum().sum())
```

<h2  style="background-color: rgb(155,196,226); color: rgb(8,69,126); padding: 10px;">Part 9. Create analysis columns</h2>
<p>This is where the project gets interesting!  Let's measure L2 language improvement.</p>
<h3  style="background-color: white; color: rgb(8,69,126); padding: 10px;">9a. L2 improvement</h3>

```python
df["absolute_gain"] = df["l2_post_score"] - df["l2_pre_score"]

df["percent_gain"] = np.where(
    df["l2_pre_score"] > 0,
    ((df["l2_post_score"] - df["l2_pre_score"]) / df["l2_pre_score"]) * 100,
    np.nan
)
display(df["percent_gain"])
```

<hr />
<h3 style="background-color: white; color: rgb(8,69,126); padding: 10px;">9b.  Cost per student</h3>

```python
df["cost_per_student"] = np.where(
    df["avg_enrollment"] > 0,
    df["cost_per_annum"] / df["avg_enrollment"],
    np.nan
)
display(df["cost_per_student"])
```

<hr/><h3 style="background-color: white; color: rgb(8,69,126); padding: 10px;">9c. Cost-effectiveness</h3>

```python
df["cost_effectiveness"] = np.where(
    df["cost_per_student"] > 0,
    df["percent_gain"] / df["cost_per_student"],
    np.nan
)
display(df["cost_effectiveness"])
```

<hr />
<h3  style="background-color: white; color: rgb(8,69,126); padding: 10px;">9d. Success threshold</h3>
<p>Now we get to answer our question.  I opted for a 15% improvement in L2 as being the metric for saying the cost of the programs collectively are worth it.</p>
<p>In practice, this value would come probably from some professional standard or some government's goals.</p>

```python
SUCCESS_THRESHOLD = 15

df["meets_success_target"] = df["percent_gain"] >= SUCCESS_THRESHOLD
```

<h2  style="background-color: rgb(155,196,226); color: rgb(8,69,126); padding: 10px;">Part 10. Research questions</h2>
<p>We started with an idea (aka the problematic), then turned it into something for a research project 
(aka the Research Question, RQ), which in turn was "operationalized" through a small number of sub-questions that collectively will feed into our understanding of the RQ.</p>

<h3  style="background-color: rgb(8,69,126); color: rgb(155,196,226); padding: 10px;">Question 1: What is the averatge starting L2 skill by language?</h3>

```python
avg_pre_by_language = (
    df.groupby("language")["l2_pre_score"]
    .mean()
    .sort_values(ascending=False)
)

display(avg_pre_by_language)
```

<h3  style="background-color: rgb(8,69,126); color: rgb(155,196,226); padding: 10px;">Question 2: Which languages/programs show the largest average improvement?</h3>

```python
avg_gain_by_language = (
    df.groupby("language")["percent_gain"]
    .mean()
    .sort_values(ascending=False)
)

display(avg_gain_by_language)
```

<h3  style="background-color: rgb(8,69,126); color: rgb(155,196,226); padding: 10px;">Question 3: What is the average cost per student by language?</h3>

```python
avg_cost_by_language = (
    df.groupby("language")["cost_per_student"]
    .mean()
    .sort_values()
)

display(avg_cost_by_language)
```

<h3 style="background-color: rgb(8,69,126); color: rgb(155,196,226); padding: 10px;">Question 4: Do K-12 and university programs differ in success?</h3>

```python
k12_avg_gain = df[df["k_12"] == 1]["percent_gain"].mean()
university_avg_gain = df[df["university"] == 1]["percent_gain"].mean()

print("K-12 average percent gain:", round(k12_avg_gain, 2))
print("University average percent gain:", round(university_avg_gain, 2))
```

```python
# alternatively
level_comparison = pd.DataFrame({
    "group": ["K-12", "University"],
    "avg_percent_gain": [
        df[df["k_12"] == 1]["percent_gain"].mean(),
        df[df["university"] == 1]["percent_gain"].mean()
    ]
})

display(level_comparison)
```

<h3 style="background-color: rgb(8,69,126); color: rgb(155,196,226); padding: 10px;">Question 5: Which programs are the most cost-effective?</h3>

```python
cost_effective_summary = (
    df.groupby("language")["cost_effectiveness"]
    .mean()
    .sort_values(ascending=False)
)

display(cost_effective_summary)
```

<h3  style="background-color: rgb(8,69,126); color: rgb(155,196,226); padding: 10px;">Question 6: Which programs meet the success target most often?</h3>
<p>This should demonstrate that each question now has a measurable variable.  This is what 
"operationalizing a research question" means.</p>

```python
success_summary = (
    df.groupby("language")
    .agg(
        avg_percent_gain=("percent_gain", "mean"),
        avg_cost_per_student=("cost_per_student", "mean"),
        pct_meeting_target=("meets_success_target", "mean")
    )
)

success_summary["pct_meeting_target"] = success_summary["pct_meeting_target"] * 100
success_summary = success_summary.sort_values("pct_meeting_target", ascending=False)

display(success_summary)
```

<p>Right now our dataset has individual program observations. 
But a policymaker or researcher usually wants summaries.  These are our actionable recommendations...</p>
<ul><li>Which languages show the most improvement?</li>
<li>Which programs are most cost-effective?</li>
<li>Do K-12 programs perform differently than university programs?</li></ul>
<p>To answer those questions, we first create new columns, then we summarize with <code>groupby()</code>.</p>

```python
# derived columns: L2 improvement
df["absolute_gain"] = df["l2_post_score"] - df["l2_pre_score"]

df["percent_gain"] = (
    (df["l2_post_score"] - df["l2_pre_score"]) /
    df["l2_pre_score"]
) * 100

# cost per student
df["cost_per_student"] = df["cost_per_annum"] / df["avg_enrollment"]

# cost effectiveness metric
df["cost_effectiveness"] = df["percent_gain"] / df["cost_per_student"]

# ------------- using agg()
language_summary = (
    df.groupby("language")
      .agg(
          avg_percent_gain=("percent_gain", "mean"),
          avg_cost_per_student=("cost_per_student", "mean"),
          avg_pre_score=("l2_pre_score", "mean"),
          avg_post_score=("l2_post_score", "mean")
      )
      .sort_values("avg_percent_gain", ascending=False)
)

display(language_summary)
```

<h2>Used in policy ... </h2>

```python
# before grouping Compare k12 - Univ.
k12_summary = df[df["k_12"] == 1]["percent_gain"].mean()
university_summary = df[df["university"] == 1]["percent_gain"].mean()

print("Average K-12 improvement:", round(k12_summary,2))
print("Average University improvement:", round(university_summary,2))

# BETTER version
education_summary = pd.DataFrame({
    "level": ["K-12", "University"],
    "avg_percent_gain": [
        df[df["k_12"] == 1]["percent_gain"].mean(),
        df[df["university"] == 1]["percent_gain"].mean()
    ]
})

display(education_summary)
```

```python
# back to cost effectiveness by language: most improvement per dollar (euro) spent:
cost_effectiveness_summary = (
    df.groupby("language")["cost_effectiveness"]
      .mean()
      .sort_values(ascending=False)
)

display(cost_effectiveness_summary)
```

```python
# a final summary report - for the client / audience
program_report = (
    df.groupby("language")
      .agg(
          avg_percent_gain=("percent_gain","mean"),
          avg_cost_per_student=("cost_per_student","mean"),
          programs=("program_id","count")
      )
      .sort_values("avg_percent_gain",ascending=False)
)

display(program_report)
```

```python
# NOTE: scatterplots can be noisy ... so summarizing helps.
summary = (
    df.groupby("language")
      .agg(
          avg_percent_gain=("percent_gain","mean"),
          avg_cost=("cost_per_student","mean")
      )
)
plt.figure(figsize=(8,6))

plt.scatter(
    summary["avg_cost"],
    summary["avg_percent_gain"]
)

for lang in summary.index:
    plt.text(
        summary.loc[lang,"avg_cost"],
        summary.loc[lang,"avg_percent_gain"],
        lang,
        fontsize=9
    )

plt.xlabel("Average Cost per Student")
plt.ylabel("Average Percent Gain")
plt.title("Cost vs Improvement by Language Program")

plt.show()
```
